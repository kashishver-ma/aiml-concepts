```python
dct = {
  'First Owner':1, 'Second Owner':2, 'Third Owner':3,
       'Fourth Owner Or More':4 
}
```

```python
df['owner']=df['owner'].map(dct)
```

```python
brand_dct = {
    'TVS':1, 'Royal Enfield':2, 'Triumph':3, 'Yamaha':4, 'Honda':5, 'Hero':6,
       'Bajaj':7, 'Suzuki':8, 'Benelli':9, 'KTM':10, 'Mahindra':11, 'Kawasaki':12,
       'Ducati':13, 'Hyosung':14, 'Harley-Davidson':15, 'Jawa':16, 'BMW':17, 'Indian':18,
       'Rajdoot':19, 'LML':20, 'Yezdi':21, 'MV':22, 'Ideal':23
}
```

```python
df['brand']=df['brand'].map(brand_dct)
```

```python
df
```

```python

x = df.drop('price', axis=1)
y = df[['price']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y ,random_state=42, test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
lr.predict(x_test)
```

```python
#how to save this model and how to use this model
#joblib already install in system

import joblib
#object nm , "model.lb" -------> saved model
model = joblib.dump(lr, "model.lb") 
```

```python
#using model direclty

model = joblib.load("model.lb")
model.predict(x_test)
```

```python
# loss=actual-predeiction
# 4 types - msc,mae,rms,rmse root mean mean square error, 

#suport vector regression

from sklearn.svm import SVR

sv=SVR()

sv.fit(x_test,y_test)
pred= sv.predict(x_test)
pred

```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
print(rf.score(x_train,y_train))
print(rf.score(x_test,y_test))
rf.predict(x_test)
# --- IGNORE ---
```
