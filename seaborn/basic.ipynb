```python
#
!pip install seaborn
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# plot in mlt and lineplot in sns
# hist in mlt  and histplt in sns 
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
plt.plot(data['total_bill'])
```

```python
sns.lineplot(data['total_bill'])
```

```python
sns.scatterplot(x=data['total_bill'],y=data['tip'])
```

```python
sns.histplot(data["total_bill"])
```

```python
plt.hist(data["total_bill"])
```

```python
# catplot and relplot

sns.relplot(data["total_bill"])
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue='time',data=data,kind='scatter',col='day')
```

```python
sns.catplot(x="day",y="total_bill", kind="strip",data=data)
```

```python
sns.catplot(x="day",y="total_bill", kind="box",data=data)
```

```python
sns.catplot(x="day",y="total_bill", kind="bar",data=data)
```

```python
sns.catplot(kind="violin",x="day",y="total_bill",data=data)
```

```python
sns.catplot(kind="boxen",x="day",y="total_bill",data=data)
```

```python
sns.countplot(x="day",data=data)
```

```python
sns.boxplot(y="total_bill",data=data)
```
